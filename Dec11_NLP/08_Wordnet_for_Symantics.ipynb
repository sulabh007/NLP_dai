### Antonyms or Opposite words

```python
ant = set()
for synset in wordnet.synsets('pass'):
    for lemma in synset.lemmas():
        if lemma.antonyms(): # When antonyms are available, add
            ant.add(lemma.antonyms()[0].name())
print('Antonyms :', ant)
```
