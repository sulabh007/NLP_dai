```python
import torch
import cv2
# Image Processing
import numpy as np
import matplotlib.pyplot as plt
```

```python
# YOLOv5 model and utilites
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained = True)
# Load pre-trained YOLOv5s model
```

```python
image = cv2.imread("jungle.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
results = model(image)
```

```python
results.render()
# Display image with predicted bounding boxes and labels
plt.imsave('abc.jpg', np.squeeze(results.render()))
# Display using matplotlib
```

```python
results.show()
```
