Relative Cropping

```python
height, width, _ = rgb_img.shape
```

```python
height
```

```python
width
```

```python
height * 0.15, width * 0.15
```

```python
height * 0.85, width * 0.85
```

```python
startRow = int(height* .15)
startCol = int(width* .15)
endRow = int(height* .85)
endCol = int(width* .85)
```

```python
plt.imshow(rgb_img[startRow:endRow,startCol:endCol])
```

```python
new_img =  img[startRow:endRow,startCol:endCol]
```

```python
plt.figure(figsize=(16,9))
plt.subplot(1,2,1)
plt.grid()
plt.imshow(rgb_img)
plt.title("Original Image" + str(rgb_img.shape))
plt.subplot(1,2,2)
plt.title('Cropped Image' + str(new_img.shape) )
plt.imshow(new_img)
```
