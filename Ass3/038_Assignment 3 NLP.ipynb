#1. Apply chuncking and show the chunks of type &lt;VBD&gt;&lt;DT&gt;

```python
grammar = "VP: {<VBD><DT>}"
chunk_parser = RegexpParser(grammar)
tags = pos_tag(word_tokenize(text))
tree = chunk_parser.parse(tags)
tree
```

#2. Find all the named entities present in article

```python
nlp = spacy.load('en_core_web_sm')
```

```python
sent = nlp(text)
for ent in sent.ents:
    print(ent.text,'\t\t',ent.label_)
```

```python
displacy.render(sent, style="ent", jupyter=True)
```

#3. Access all the verbs and apply the porter stemmer and lancaster stemmer

```python
tags = pos_tag(word_tokenize(text))
tags
```

```python
verbs = set()
for i in tags:
    if i[1].startswith('V') and i[0].isalpha() :
        verbs.add(i[0])
print(verbs)
```

```python
ps = PorterStemmer()
for verb in verbs:
    print(verb,' - ',ps.stem(verb))
```

```python
ls = LancasterStemmer()
for verb in verbs:
    print(verb,' - ',ls.stem(verb))
```

#4. Access all the verbs in past tense and lemmatize them. Show only unique verbs.

```python
wnl = WordNetLemmatizer()
verbs = set()
for i in tags:
    if i[1].startswith('VBD') and i[0].isalpha() :
        verbs.add(i[0])
        
for word in verbs:
    print(word,' - ',wnl.lemmatize(word, pos = 'v'))
```

#5. Read only the adjectives from the text and find their synonyms.

```python
adj = set()
for i in tags:
    if i[1].startswith('J') and i[0].isalpha() :
        adj.add(i[0])
print(adj)
```

```python
for word in adj:
    syns = set()
    for syn in wordnet.synsets(word):
        syns.add(syn.lemmas()[0].name())
    syns.discard(word)
    if len(syns):
        print("Synonyms of", word, 'are:', syns)
```

#6. Read all the verbs and print their antonyms.

```python
verbs = set()
for i in tags:
    if i[1].startswith('V') and i[0].isalpha() :
        verbs.add(i[0])
```

```python
for word in verbs:
    ant = []
    for synset in wordnet.synsets(word):
        for lemma in synset.lemmas():
            if lemma.antonyms():
                ant.append(lemma.antonyms()[0].name())
    if len(ant):
        print("Antonyms of", word, 'are:', ant)
```

#7. Print all the entities marked as DATE and TIME in the data

```python
sent = nlp(text)
for ent in sent.ents:
    if ent.label_=='DATE' or ent.label_=='TIME':
        print(ent.text,'\t:\t',ent.label_)
```

#8. Print all the named of locations present in the article.

```python
loc = set()
for ent in sent.ents:
    if ent.label_=='GPE' or ent.label_=='LOC':
        loc.add(ent.text)
loc
```
