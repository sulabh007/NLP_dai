```python
import matplotlib.pyplot as plt 
import numpy as np 
from PIL import Image
import cv2
```

```python
img = cv2.imread('dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

#### Extract interest points

```python
# convert the image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# create the object of SIFT algorithm
sift = cv2.SIFT_create()

# Detect the compute the keypoints
kp, _ = sift.detectAndCompute(gray, None)

# draw keypoints on the image and highlight
new_img = cv2.drawKeypoints(gray,kp,img,flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.imshow(new_img)
```
