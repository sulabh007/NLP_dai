```python
import pandas as pd
import numpy as np
import os
```

```python
path = os.getcwd()
```

```python
headlines = []
for filename in os.listdir(path):
  if "Articles" in filename:
    article_df = pd.read_csv(path+ '/' + filename)
    headlines.extend(list(article_df['headline'].values))
```

```python
len(headlines)
```

```python
headlines
```

```python
headlines = [ h for h in headlines if h != 'Unknown']
print("The number of headlines is : ", len(headlines))
```

```python
import string
import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter(action = 'ignore', category= FutureWarning)
```

```python
def clean_text(headline):
  text = "".join( word for word in headline if word not in string.punctuation).lower()
  text = text.encode('utf8').decode('ascii', 'ignore')
  return text

corpus = [ clean_text(headline) for headline in headlines]
```

```python
corpus
```

```python
vocab = []
for line in corpus:
  words = line.split()
  for word in words:
    vocab.append(word)

vocabulary = set(vocab)

len(vocabulary)
```

```python
from keras.preprocessing.text import Tokenizer
from keras_preprocessing.sequence import pad_sequences
from keras.utils import to_categorical
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
word2index = tokenizer.word_index
len(word2index)
```

```python
word2index
```

```python
dictionary = {}
rev_dictionary = {}
for word, idx in word2index.items():
  dictionary[word] = idx
  rev_dictionary[idx] = word
```

```python
dictionary, rev_dictionary
```

```python
input_sequences = tokenizer.texts_to_sequences(corpus)
```

```python
input_sequences
```

```python
input_data = []
target = []
for line in input_sequences:
  for i in range(1, len(line)-1):
    input_data.append(line[:i])
    target.append(line[i])
```

```python
input_data
```

```python
input_data[0], target[0]
```

```python
input_data[1], target[1]
```

```python
len(input_data)
```

```python
MAX_LEN = 0
for seq in input_data:
  if len(seq) > MAX_LEN:
    MAX_LEN = len(seq)

MAX_LEN
```

```python
input_data = pad_sequences(input_data, maxlen = MAX_LEN, padding = 'post')

input_data
```

```python
input_data.shape
```

```python
target = to_categorical(target)
```

```python
target
```

```python
target.shape
```

```python
from keras.models import Sequential
from keras.layers import Embedding, GRU, Dense, Dropout, LSTM
from keras.callbacks import EarlyStopping
```

```python
model = Sequential()

model.add(Embedding(input_dim = len(vocabulary), output_dim = 100, input_length= MAX_LEN))

model.add(LSTM(units= 100))

# model.add(Dropout(rate = 0.1))

model.add(Dense(units = target.shape[1], activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(input_data, target, batch_size=10, epochs=10)
```

```python
new_sent = 'Trump'
```

```python
def predict(new_sent):
  new_sent = clean_text(new_sent)
  new_seq = tokenizer.texts_to_sequences([new_sent])
  new_seq = pad_sequences(new_seq, maxlen = MAX_LEN, padding = 'post')
  preds = model.predict(new_seq, verbose = False)
  return rev_dictionary[np.argmax(preds)]
```
