Prepare the data

```python
alphabets = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
len(alphabets)
```

Preprocess the data

```python
int_to_char = dict(enumerate(alphabets))
int_to_char
```

```python
char_to_int = dict((v,i) for i, v in enumerate(alphabets))
char_to_int
```

```python
int_to_char[5]
```

```python
char_to_int['H']
```

```python
seq_length = 1
```

```python
x = []
y = []
for i in range(len(alphabets)-seq_length):
    seq_in = [alphabets[i]]
    seq_out = alphabets[i+1]
    x += [[char_to_int[seq_in[0]]]]
    y += [char_to_int[seq_out]]
    print(seq_in, '->', seq_out)
```

```python
x
```

```python
y
```

#### One hot encoding

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_new = to_categorical(y)
```

```python
y_new
```

```python
y_new.shape
```

```python
import numpy as np
```

```python
np.array(x).shape
```

```python
x = np.reshape(x, (25,1,1))
x.shape
```

```python
x
```

```python
x = x / 25
x
```

#### Build the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(128))
model.add(Dense(26, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x,y_new, epochs=100, batch_size=1)
```

```python
model.evaluate(x, y_new, batch_size=1)
```

#### Predict on the new data

```python
new = 'A'
a = char_to_int[new]
a = np.array(a)
a = np.reshape(a, (1,1,1))
a.shape

x_test = a / 25
x_test

pred = model.predict(x_test)
pred = pred.argmax(axis=1)
int_to_char[pred[0]]
```

```python
def predict(char):
    a = char_to_int[char]
    a = np.array(a)
    a = np.reshape(a, (1,1,1))
    x_test = a / 25
    pred = model.predict(x_test,verbose=0)
    pred = pred.argmax(axis=1)
    return int_to_char[pred[0]]
```
